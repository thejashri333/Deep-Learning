```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
breast_cancer_data = load_breast_cancer()
print(breast_cancer_data)
```

```python
data_frame = pd.DataFrame(breast_cancer_data.data, columns=breast_cancer_data.feature_names)
```

```python
data_frame['Label']= breast_cancer_data.target
```

```python
X= data_frame.drop(columns='Label', axis=1)
Y=data_frame['Label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train , X_test, Y_train , Y_test = train_test_split(X,Y,test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape,X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import  StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

Building the Neural Network

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up  layers  of  neural network 

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(3, activation='sigmoid')
])
```

```python
# Compilingthe Neural Network

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',
metrics=['accuracy'])
```

Trainning  the Neural Network

```python
history = model.fit(X_train_std, Y_train, validation_split=0.1 , epochs= 10)
```

Viusalizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model  accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data', 'validation data'], loc='lower right')
```

Accuracy of the model on test data

```python
loss , accuracy = model.evaluate(X_test_std,Y_test)
print(accuracy)
```
